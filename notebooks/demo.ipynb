## ğŸ§ª Demo 1: Code Execution Sandbox

Safe code execution is crucial for hands-on learning.


````python
from core.code_sandbox import execute_code

# Example: Execute simple Python code
code = """
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

for i in range(8):
    print(f"F({i}) = {fibonacci(i)}")
"""

result = execute_code(code)
print(f"âœ… Success: {result.success}")
print(f"â±ï¸ Execution Time: {result.execution_time:.3f}s")
print(f"\nğŸ“¤ Output:\n{result.output}")
````

## ğŸ“Š Demo 2: Learning Progress Tracking

The system tracks learning metrics to personalize the experience.


````python
from core.memory import Memory, UserProfile

# Create a user profile
memory = Memory()
profile = UserProfile(
    user_id="demo_learner",
    learning_style="visual",
    pace_preference="moderate",
    current_language="python",
    current_topic="loops",
    accessibility_needs=["adhd"]
)
memory.set_user_profile(profile)

# Simulate learning activity
memory.update_learning_metric("loops", success=True, practice_time=120)
memory.update_learning_metric("loops", success=True, practice_time=90)
memory.update_learning_metric("functions", success=False, practice_time=150)
memory.update_learning_metric("lists", success=True, practice_time=100)

# View progress
print("ğŸ“ˆ Learning Metrics:")
print("=" * 50)
for topic, metric in memory.get_all_metrics().items():
    print(f"\nğŸ“š {topic.upper()}")
    print(f"   Skill Level: {'â­' * int(metric.skill_level * 5)} ({metric.skill_level:.2f})")
    print(f"   Practice Count: {metric.practice_count}")
    print(f"   Success Rate: {metric.success_rate:.0%}")

print(f"\n\nğŸ’ª Strong Topics: {', '.join(memory.get_strong_topics())}")
print(f"ğŸ¯ Need Practice: {', '.join(memory.get_weak_topics())}")
````

## ğŸ¤– Demo 3: Multi-Agent System in Action

See how different agents handle different types of requests.


````python
from core.llm_service import create_llm_service
from core.code_sandbox import CodeSandbox
from agents.orchestrator import Orchestrator

# Initialize the system
llm_service = create_llm_service()
code_sandbox = CodeSandbox()
orchestrator = Orchestrator(llm_service, memory, code_sandbox)

print("âœ… CodeMentor AI System Initialized!\n")
````

````python
# Example 1: Learning Request (Tutor Agent)
print("ğŸ‘¨â€ğŸ« TUTOR AGENT - Explaining Concept\n" + "=" * 60)
response = orchestrator.chat("Explain what a Python list is in simple terms")
print(response)
print("\n")
````

````python
# Example 2: Debugging Request (Debug Agent)
print("ğŸ”§ DEBUG AGENT - Analyzing Error\n" + "=" * 60)

buggy_code = """
def calculate_average(numbers):
    total = 0
    for num in numbers:
        total += num
    return total / len(numbers)

scores = [85, 90, 78, 92]
average = calculate_average(scores)
print(f"Average: {average}")
"""

debug_response = orchestrator.chat(
    f"My code has an error. Can you help? \n```python\n{buggy_code}\n```"
)
print(debug_response)
print("\n")
````

````python
# Example 3: Assessment Request (Assessment Agent)
print("ğŸ“ ASSESSMENT AGENT - Creating Quiz\n" + "=" * 60)
quiz = orchestrator.create_quiz("Python lists", difficulty="beginner")
print(quiz)
print("\n")
````

````python
# Example 4: Motivation Request (Motivation Agent)
print("ğŸŒŸ MOTIVATION AGENT - Progress Update\n" + "=" * 60)
progress = orchestrator.get_progress_update()
print(progress)
print("\n")
````
